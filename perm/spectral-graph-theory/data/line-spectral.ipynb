```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.diag(np.ones(20), 1)
```

```python
a = a + a.T
```

```python
la = np.diag(a.sum(axis=0)) - a
```

```python
d,v = np.linalg.eigh(la)
```

```python
plt.plot(v[:, 0])
```
